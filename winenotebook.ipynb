```python
#SETTING OUR DATA AND lABELS
```

```python
x=wd.drop("Wine",axis=1)
```

```python
Y=wd["Wine"]
```

```python
#splitting our data
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,Y,test_size=.20)
```

```python
#Data preprocessing
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ms=StandardScaler()
```

```python
ms.fit(X_train)
```

```python
X_train=ms.transform(X_train)
```

```python
type(X_train)
```

```python
#training the model
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
mlp=MLPClassifier(hidden_layer_sizes=(13,13,13),max_iter=500)
```

```python
#training with data
```

```python
mlp.fit(X_train,y_train)
```

```python
mm=mlp.fit(X_train,y_train)
```

```python
#predicition and evaluation
```

```python
nm=mm.predict(X_test)
```

```python
nm
```

```python
from sklearn import metrics
```

```python
metrics.accuracy_score(y_test,nm)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
confusion_matrix(y_test,nm)
```

```python
classification_report(y_test,nm)
```
